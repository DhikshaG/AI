```python
# Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/playtennis.csv')

# Prepare the data
X = df.drop('Play Tennis', axis=1)
y = df['Play Tennis']

# Convert categorical columns to numeric using one-hot encoding
X = pd.get_dummies(X)

# Build the Decision Tree
model = DecisionTreeClassifier(criterion='gini', random_state=0)  # You can change to 'entropy' for ID3
model.fit(X, y)

# Visualize the Decision Tree
plt.figure(figsize=(20,10))
plot_tree(model, filled=True, feature_names=X.columns, class_names=model.classes_)
plt.show()
```
